Replace `Cyg X-1` below with your source of choice and tweak `search_radius` to your preference (in arcsec).

If the cell runs without any issue and the source is correctly resolved and you see only 1 source found in DR2 and DR3, it should run without any problem. If the number of sources found is not 1, tweak the search radius so that it becomes 1 for both.

In the cell below, `tau_dr2` and `tau_dr3` are cutoff thresholds (in probability density) for determining the confidence interval on the distance PDF. You can tweak them to achieve the desired confidence interval (which you can check printed in the outputs).

`distance_range` is simply the range of distances to plot for the PDF in kpc.

```python
source_name = 'Cyg X-1'
search_radius = 0.8         # in arcsec
distance_range = [1.5,3.5]  # range to plot in kpc

# Advanced input:
distance_prior = 'both'   # choice of prior between Bailer-Jones+18 ("bj") and Atri+19 ("atri"), or both ("both")
tau_dr2 = 1.58            # Probability cut for the Atri prior in DR2 (not used for BJ prior)
tau_dr3 = 3.5             # Probability cut for the Atri prior in DR3 (not used for BJ prior)

tab, fig = edr3funcs.gaiaedr3_plots(source_name, search_radius, tau_dr2, tau_dr3, distance_range, distance_prior)

# If you want to search with coordinates instead of name:
# tab, fig = edr3funcs.gaiaedr3_plots_coords('19 58 21.6758', '+35 12 05.782', search_radius, tau_dr2, tau_dr3, distance_range, distance_prior)

fig
tab
```
